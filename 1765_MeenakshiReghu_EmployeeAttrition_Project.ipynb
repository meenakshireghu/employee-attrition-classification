### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
df['Attrition']=df['Attrition'].map({'Yes':1,'No':0})
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
new_df=pd.get_dummies(df,drop_first=True)
new_df.head()
```

#### Check the shape of our new dataset

```python
new_df.shape
```

#### Print unique values in our new dataframe

```python
for col in df:
    print(f"Unique values in {col}:{df[col].unique()}")
```

#### Split the columns into input and target variables (X and y)

```python
X=new_df.drop(columns=['Attrition'])
y=new_df['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
xcolumns=X.columns
X = scaler.fit_transform(X)
X=pd.DataFrame(X,columns=xcolumns)
```

```python
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model_score=model.score(X_train,y_train)
print("Accuracy score of our model is :",model_score)
print("Confusion matrix is:\n",metrics.confusion_matrix(y_test,y_pred))
print("Classification report is:\n",metrics.classification_report(y_test,y_pred))
score=cross_val_score(model,X,y,cv=5)
mean_score_log=score.mean()
print("Mean of Cross validation score is: ",mean_score_log)
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred),annot=True,fmt='d')
plt.show()
```

### Random Forest Classifier
** Choose the best estimator and parameters :GridSearchCV**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params = {'max_depth': [0.5, 1, 5, 10],
          'n_estimators': [16, 32, 50, 100], 
          'n_jobs': [1, 2],
          'random_state': [1, 10, 20, 42]}
```

```python
gsv = GridSearchCV(RandomForestClassifier(), params,cv = 5, scoring = 'accuracy')
```

```python
gsv.fit(X_train,y_train)
```

```python
print("Best estimator is: ",gsv.best_estimator_)
print("Best parameters are: ",gsv.best_params_)
print("Best score is: ",gsv.best_score_)
```

<span style = "color:blue"> Create Random forest model with the best parameters </span>

```python
new_model =RandomForestClassifier(max_depth=10, n_estimators=32, n_jobs=1, random_state=20)
```

```python
new_model.fit(X_train,y_train)
```

```python
y_pred = new_model.predict(X_test)
```

```python
ran_model_score=new_model.score(X_train,y_train)
print("Accuracy score of our model is :",ran_model_score)
print("Confusion matrix is:\n",metrics.confusion_matrix(y_test,y_pred))
print("Classification report is:\n",metrics.classification_report(y_test,y_pred))
score=cross_val_score(new_model,X,y,cv=5)
mean_score_ran=score.mean()
print("Mean of Cross validation score is: ",mean_score_ran)
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
plt.figure(figsize=(6,3))
sns.heatmap(metrics.confusion_matrix(y_test,y_pred),annot=True,fmt='d')
plt.show()
```
